## Visualize Class Activations

### Extract and Plot Classmaps

We now explore the localization ability of the pre-trained ResNet-50 model.  The localization is expressed as a heat map (henceforth referred to as a class activation map), where the color-coding scheme identifies regions that are relatively important for the GAP-CNN to perform the object identification task.  Some sample output from the [first paper](http://cnnlocalization.csail.mit.edu/Zhou_Learning_Deep_Features_CVPR_2016_paper.pdf) to introduce this technique appears below, where the authors were able to train a model to determine regions of an image that correspond to humans performing different actions.  

<img src="images/sample_GAPCNN_output.png" width="400">

We won't into the details of how the class activation map works, but it is well-explained in [this paper](http://cnnlocalization.csail.mit.edu/Zhou_Learning_Deep_Features_CVPR_2016_paper.pdf).  The main idea is that each of the filtered images in the final convolutional output acts as a detector for a different pattern in the image, localized in space (with detected patterns manifest in the corresponding filtered image as larger pixel values).  To get the class activation map, we need only to translate these detected patterns to detected objects.  This translation is done by noticing each GAP output corresponds to a different filtered image, and that the weights in the final dense layer encode each filtered image's contribution to the predicted object class.  We can thus sum the contributions of each of the detected patterns (from the final convolutional output), where detected patterns that are more important to the predicted object class are given more weight.  

We have extensively annotated our code that produces the class activation map, which appears below in the function `plot_classmap`.  Let $f_k$ represent the $k$-th filtered image in the final convolution output.  In the pre-trained ResNet-50 model, the last convolutional output contains 2048 filtered images, each 7 pixels high and 7 pixels wide.  So, $f_0$ is $7\times7$ pixels and is the first filtered image in the convolutional output, $f_1$ is also $7\times7$ pixels and is the second filtered image, and so on, where $f_{2047}$ is likewise $7\times7$ pixels and is the final filtered image.  In order to permit comparison to the original image, we use bilinear upsampling to resize each filtered image to $224 \times 224$.

Next, we look at the class that is predicted by the model.  The output node corresponding to the predicted class is connected to every node in the previous (GAP) layer.  Let $w_i$ represent the weight connecting the $i$-th node in the GAP layer to the output node corresponding to the predicted dog breed.  Then, in order to obtain the class activation map, we need only compute the sum

$$
w_0 \cdot f_0 + w_1 \cdot f_1 + \ldots + w_{2047} \cdot f_{2047}.
$$

If none of that makes sense to you, don't worry!  You will only need to execute the implementation that we have written for you in the `plot_many_classmaps` function below.  We explain how to do this below the next code block.

We can now use the `plot_many_classmaps` function to plot the class activation maps for the pretrained ResNet-50 model.  To use the function, you first need to create a figure using the `matplotlib` library.  Then you can execute the `plot_many_classmaps` function, which takes as input both the figure and a numpy array with filenames linking to images.  The function returns the class activation maps and original images corresponding to the first five filenames in the input array.

````python
import cv2                                             # for detecting + segmenting human faces
import h5py                                            # for loading pre-trained model files
import matplotlib.pyplot as plt                        # make plots
%matplotlib inline                               
import numpy as np                                     # scientific computing
import os                                              # extract file names and check if files exist
import scipy                                           # bilinear upsampling
from tqdm import tqdm                                  # time algorithms
import ast                                             # importing imagenet classes dictionary
from sklearn.model_selection import train_test_split   # for splitting data into train, test, validation sets
from sklearn.datasets import load_files                # for importing dog dataset from directory structure
from PIL import ImageFile                              # load big files
ImageFile.LOAD_TRUNCATED_IMAGES = True                 # load big files

from keras.applications.vgg16 import VGG16
from keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
from keras.models import Model                         # construct new models
from keras.preprocessing import image                  # import images in Keras
from keras.layers import GlobalAveragePooling2D, Dense # for constructing model
from keras.callbacks import ModelCheckpoint            # for model checkpointing
````

````python
##########################
# Import human filenames #
##########################

# dataset lives here: http://vis-www.cs.umass.edu/lfw/#download
human_filenames = ['humanImages/'+f for f in os.listdir('humanImages/') if f.endswith('.jpg')]
print('There are %d total human images.' % len(human_filenames))

######################
# Import dog dataset #
######################

# dataset lives here: http://www.umiacs.umd.edu/~kanazawa/
# start import with sklearn
dog_dataset = load_files('dogImages')
dog_filenames = np.array(dog_dataset['filenames'])
dog_targets = np.array(dog_dataset['target'])
dog_breed_names = dog_dataset['target_names']

# remove all files that are not images
bool_to_keep = np.array([filename.endswith('jpg') for filename in dog_filenames])
dog_filenames = dog_filenames[bool_to_keep]
dog_targets = dog_targets[bool_to_keep]

print('There are %d total dog categories.' % len(np.unique(dog_targets)))
print('There are %d total dog images.' % len(dog_targets))
````

````python
def path_to_tensor(img_path):
    # loads RGB image as PIL.Image.Image type
    img = image.load_img(img_path, target_size=(224, 224))
    # convert PIL.Image.Image type to 3D tensor with shape (224, 224, 3)
    x = image.img_to_array(img)
    # convert 3D tensor to 4D tensor with shape (1, 224, 224, 3) and return 4D tensor
    return np.expand_dims(x, axis=0)

def pretrained_path_to_tensor(img_path):
    # convert to Keras-friendly, 4D tensor 
    img = path_to_tensor(img_path)
    # convert RGB -> BGR, subtract mean ImageNet pixel, and return 4D tensor
    return preprocess_input(img)
````

````python
# define ResNet50 model
ResNet50_model = ResNet50(weights='imagenet')

# load the dictionary that identifies each ImageNet category to an index in the prediction vector
with open('support_files/imagenet1000_clsid_to_human.txt') as imagenet_classes_file:
    imagenet_classes_dict = ast.literal_eval(imagenet_classes_file.read())


def plot_classmap(img_path, ax): 
    # turn image into Keras-friendly tensor
    im = pretrained_path_to_tensor(img_path)
    
    # get model's prediction (number between 0 and 999, inclusive)
    pred = np.argmax(ResNet50_model.predict(im))
    
    # get weights 
    amp_layer_weights = ResNet50_model.layers[-1].get_weights()[0][:, pred] # dim: (2048,) 

    # get filtered images from convolutional output corresponding to supplied image 
    last_conv_output_model = Model(input=ResNet50_model.input, output=ResNet50_model.layers[-4].output) 
    last_conv_output = np.squeeze(last_conv_output_model.predict(im)) # dim: 7 x 7 x 2048

    # bilinear upsampling to resize each filtered image to size of original image 
    mat_for_mult = scipy.ndimage.zoom(last_conv_output, (32, 32, 1), order=1) # dim: 224 x 224 x 2048

    # get class activation map for object class that is predicted to be in the image
    final_output = np.dot(mat_for_mult.reshape((224*224, 2048)), amp_layer_weights).reshape(224,224) # dim: 224 x 224
    
    # plot class activation map
    ax.imshow(final_output, cmap='jet')
    ax.set_title(imagenet_classes_dict[pred])
    
# load image, convert BGR --> RGB, resize image to 224 x 224, and plot image
def plot_image(img_path, ax, alpha):
    im = cv2.resize(cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB), (224, 224))
    ax.imshow(im, alpha=alpha)
    
    
# plot the class activation map + original image for the first five images in supplied filenames
def plot_many_classmaps(filenames, fig):
    fig.subplots_adjust(left=0, right=1, bottom=0.4, top=1, hspace=0.05, wspace=0.05)
    for i in range(5):
        ax = fig.add_subplot(2, 5, i+1, xticks=[], yticks=[])
        plot_classmap(filenames[i], ax)
        plot_image(filenames[i], ax, 0.3)
        ax = fig.add_subplot(2, 5, i+6, xticks=[], yticks=[])
        plot_image(filenames[i], ax, 1)
````

````python
fig = plt.figure(figsize=(20, 15))
plot_many_classmaps(dog_filenames, fig)
````

### Towards State-of-the-Art 

For classification tasks, a common choice for CNN architecture is repeated blocks of convolution and maxpooling layers, followed by two or more densely connected layers, where the final dense layer has a softmax activation function and a node for each potential object category.  

As an example, consider the VGG-16 model architecture, depicted in the figure below.

![VGG-16 Model Figure][image3]

We can also summarize the layers of the VGG-16 model by executing the following line of code in the terminal:

```	
	python -c 'from keras.applications.vgg16 import VGG16; VGG16().summary()'
```

Your output should appear as follows:

![VGG-16 Model Keras Layers][image2]

You will notice five blocks of (two to three) convolutional layers followed by a max pooling layer.  The final max pooling output is then  flattened and followed by three densely connected layers.  Notice the total number of parameters in the model.  Most of these parameters belong to the fully connected layers!

As you can probably imagine, an architecture like this has the risk of overfitting to the training dataset.  In practice, judicious use of dropout laters is used to avoid overfitting.

In the last few years, experts have used global average pooling (GAP) layers to reduce the total number of parameters in the model.  The [first paper](https://arxiv.org/pdf/1312.4400.pdf) to propose GAP layers designed an architecture where the output of the final max pooling layer was a filtered image with one channel (or feature map) for each image category in the dataset.  The max pooling output was then fed to a GAP layer, which yielded a vector with a single entry for each possible object in the classification task.  The authors then applied a softmax activation function to yield the predicted probability of each class.  If you peek at the [original paper](https://arxiv.org/pdf/1312.4400.pdf), we especially recommend checking out Section 3.2, titled "Global Average Pooling".

The [ResNet-50 model](http://ethereon.github.io/netscope/#/gist/db945b393d40bfa26006), which is considered to be more state-of-the-art than the VGG-16 model, takes a less extreme approach; instead of getting rid of dense layers altogether, its final convolutional output is fed to a GAP layer, followed by one densely connected layer with a softmax activation function that yields the predicted object classes.  

To summarize, VGG-16 has __two__ hidden fully connected layers, the first paper to propose GAP layers has __zero__ fully connected layers (hidden or otherwise), and ResNet-50 has __one__ fully connected output layer and __zero__ hidden fully connected layers.  In this project, we will explore structures inspired by the ResNet-50 architecture.

In mid-2016, [researchers at MIT](http://cnnlocalization.csail.mit.edu/Zhou_Learning_Deep_Features_CVPR_2016_paper.pdf) demonstrated that CNNs with GAP layers (a.k.a. GAP-CNNs) that have been trained for a classification task can also be used for object localization.  That is, a GAP-CNN not only tells us __*what*__ object is contained in the image - it also tells us __*where*__ the object is in the image, and through no additional work on our part!
